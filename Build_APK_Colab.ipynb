```python
# 2️⃣ 下载源码
!git clone https://github.com/yadang2026/InstaGuard.git
%cd InstaGuard
print('✅ 源码下载完成')
```

```python
# 3️⃣ 修复构建配置
%%bash
sed -i 's/libncurses5-dev //' .github/workflows/main.yml 2>/dev/null || true
sed -i 's/requirements = python3,kivy.*/requirements = python3,kivy==2.3.0,https:\/\/github.com\/kivymd\/KivyMD\/archive\/refs\/heads\/master.zip,requests,httpx,lxml/' buildozer.spec
sed -i 's/# android.accept_sdk_license/android.accept_sdk_license/' buildozer.spec
sed -i 's/android.api = 34/android.api = 33/' buildozer.spec
echo '✅ 配置修复完成'
```

```python
# 4️⃣ 构建 APK（约 15-20 分钟）
!buildozer android debug 2>&1 | tail -50
```

```python
# 5️⃣ 下载 APK
from google.colab import files
import glob
apk = glob.glob('bin/*.apk')
if apk:
    print(f'✅ APK: {apk[0]}')
    files.download(apk[0])
else:
    print('❌ 未找到 APK，请检查构建日志')
```
